```python
# Drop 'abs_error' and 'y_cv_pred' from all DataFrame globals that start with 'protein_'
cols_to_drop = ['abs_error', 'y_cv_pred']

for var_name, var_val in list(globals().items()):
    if var_name.startswith('protein_') and isinstance(var_val, pd.DataFrame):
        present = [c for c in cols_to_drop if c in var_val.columns]
        if present:
            var_val.drop(columns=present, inplace=True, errors='ignore')
            print(f"Dropped {present} from {var_name}")
        else:
            print(f"No target columns in {var_name}")
```

```python
import plotly.graph_objs as go

def plot_actual_vs_predicted_plotly(model, df, target_col, title='Actual vs Predicted'):
    y_actual = df[target_col]
    X = df.drop(columns=[target_col])
    y_pred = model.predict(X)

    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=y_actual,
        y=y_pred,
        mode='markers',
        marker=dict(size=6, color='blue', opacity=0.7),
        name='Data',
        text=df.index,  # Menampilkan index pada hover
        hovertemplate='Index: %{text}<br>Actual: %{x}<br>Predicted: %{y}<extra></extra>'
    ))
    # Add y=x reference line
    min_val = min(y_actual.min(), y_pred.min())
    max_val = max(y_actual.max(), y_pred.max())
    fig.add_trace(go.Scatter(
        x=[min_val, max_val],
        y=[min_val, max_val],
        mode='lines',
        line=dict(dash='dash', color='red'),
        name='y = x'
    ))
    fig.update_layout(
        title=title,
        xaxis_title='Actual',
        yaxis_title='Predicted',
        width=600,
        height=500
    )
    import plotly.io as pio
    if not os.path.exists('plot_act_pred_plotly'):
        os.makedirs('plot_act_pred_plotly')
    pio.write_html(fig, file=f'plot_act_pred_plotly/{title}.html', auto_open=False)
    fig.show()
```
