```python
import pandas as pd


# Define file paths
gula_reduksi_file_path = '../raw_data/200_data_montecarlo_NIR_Fermentasi_Gula Reduksi.csv'
kadar_air_file_path = '../raw_data/200_data_montecarlo_NIR_Fermentasi_KadarAir.csv'
lemak_file_path = '../raw_data/200_data_montecarlo_NIR_Fermentasi_Lemak.csv'
protein_file_path = '../raw_data/200_data_montecarlo_NIR_Fermentasi_Protein.csv'
fenol_file_path = '../raw_data/200_data_montecarlo_NIR_Fermentasi_Total Fenol.csv'


# Read the CSV files
gula_reduksi_df = pd.read_csv(gula_reduksi_file_path,index_col=0)
kadar_air_df = pd.read_csv(kadar_air_file_path,index_col=0)
lemak_df = pd.read_csv(lemak_file_path,index_col=0)
protein_df = pd.read_csv(protein_file_path,index_col=0)
fenol_df = pd.read_csv(fenol_file_path,index_col=0)
```

```python
# Drop the "KATEGORI" column from all relevant DataFrames if it exists
for df in [lemak_df, protein_df,gula_reduksi_df, kadar_air_df]:
    if "KATEGORI" in df.columns:
        df.drop(columns=["KATEGORI"], inplace=True)
    if "Kategori" in df.columns:
        df.drop(columns=['Kategori'], inplace=True)
```
