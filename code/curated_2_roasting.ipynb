```python
import pandas as pd


# Define file paths
gula_reduksi_file_path = '../raw_data/MONTE CARLO DATA NIR ROASTING GULA REDUKSI.csv'
kadar_air_file_path = '../raw_data/MONTE CARLO DATA NIR ROASTING KADAR AIR.csv'
lemak_file_path = '../raw_data/MONTE CARLO DATA NIR ROASTING LEMAK.csv'
protein_file_path = '../raw_data/MONTE CARLO DATA NIR ROASTING PROTEIN.csv'
fenol_file_path = '../raw_data/MONTE CARLO DATA NIR ROASTING FENOL.csv'


# Read the CSV files
gula_reduksi_df = pd.read_csv(gula_reduksi_file_path,index_col=0)
kadar_air_df = pd.read_csv(kadar_air_file_path,index_col=0)
lemak_df = pd.read_csv(lemak_file_path,index_col=0)
protein_df = pd.read_csv(protein_file_path,index_col=0)
fenol_df = pd.read_csv(fenol_file_path,index_col=0)
```

```python
# Drop the "KATEGORI" column from all relevant DataFrames if it exists
for df in [lemak_df, protein_df,gula_reduksi_df, kadar_air_df]:
    if "KATEGORI" in df.columns:
        df.drop(columns=["KATEGORI"], inplace=True)
    if "Kategori" in df.columns:
        df.drop(columns=['Kategori'], inplace=True)
```

```python
for df in [gula_reduksi_df, kadar_air_df, lemak_df, protein_df]:
    if "Unnamed: 0" in df.columns:
        df.drop(columns=["Unnamed: 0"], inplace=True)
```
